#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

casas = pd.read_csv('assets/real_estate.csv', sep=';')
municipio=['Fuenlabrada','Leganés','Getafe','Alcorcón']
surmadrid=casas[casas['level5'].isin(municipio)]
#print(surmadrid)

fuenla=surmadrid[surmadrid['level5']=='Fuenlabrada']
fuenla['nfuenla'] = (fuenla['price']-fuenla['price'].mean())/fuenla['price'].std()
print(fuenla['nfuenla'])

leganes=surmadrid[surmadrid['level5']=='Leganés']
leganes['nlega']=(leganes['price']-leganes['price'].mean())/leganes['price'].std()
print(leganes['nlega'])

getafe=surmadrid[surmadrid['level5']=='Getafe']
getafe['ngeta']=(getafe['price']-getafe['price'].mean())/getafe['price'].std()
print(getafe['ngeta'])

alcorcon=surmadrid[surmadrid['level5']=='Alcorcón']
alcorcon['nalcor']=(alcorcon['price']-alcorcon['price'].mean())/alcorcon['price'].std()
print(alcorcon['nalcor'])

convalor=pd.concat([fuenla['nfuenla'],leganes['nlega'],getafe['ngeta'],alcorcon['nalcor']])
plt.hist(convalor, bins=20, alpha=0.7)
plt.title('Precio por población (normalizado)')
plt.xlabel('Precio normalizado')
plt.ylabel('Precio por poblacion')
plt.show()

# min-max: df['columna_normalizada'] = (df['columna'] - df['columna'].min()) / (df['columna'].max() - df['columna'].min())
# z: df['columna_estandarizada'] = (df['columna'] - df['columna'].mean()) / df['columna'].std()
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

casas = pd.read_csv('assets/real_estate.csv', sep=';')
municipio=['Fuenlabrada','Leganés','Getafe','Alcorcón']
surmadrid=casas[casas['level5'].isin(municipio)]
#print(surmadrid)

fig, ax = plt.subplots(figsize = (10,6))
for poblacion in municipio:
    subset = surmadrid[surmadrid['level5'] == poblacion]
    ax.scatter(subset.index, subset['price'], label=poblacion)

ax.set_xlabel('Indice')
ax.set_ylabel('Precio')
ax.set_title('Precio de viviendas por poblacion')
ax.legend()
plt.show()
```

```python
## HERE: plot the coordinates of the estates
from ipyleaflet import Map, Marker
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

casas = pd.read_csv('assets/real_estate.csv', sep=';')
municipio=['Fuenlabrada','Leganés','Getafe','Alcorcón']
surmadrid=casas[casas['level5'].isin(municipio)]
coordinate=surmadrid[surmadrid['level5'].isin(municipio)][['latitude', 'longitude']]

## PUT HERE YOUR CODE:
for index, row in coordinate.iterrows():
    latitud, longitud = float(row['latitude'].replace(',', '.')),float (row['longitude'].replace(',', '.'))
    marker = Marker(location = (latitud, longitud))
    map.add_layer(marker)
map
```
